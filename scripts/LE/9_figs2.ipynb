```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd
import os
import glob

plt.rcParams["font.family"] = "DejaVu Serif"
plt.rcParams['font.size'] = 12
plt.rcParams['figure.dpi'] = 300
plt.rcParams['axes.linewidth'] = 2
plt.rcParams['font.weight'] = "bold"
plt.rcParams['legend.framealpha'] = 0

x = sorted(glob.glob('pickles/**.pkl'))
for q in x:
    print(q)
```

```python
# all strings
param = 'nn_hid_lay_size'
value = ['4','30','200','1000','dcrrnn']
neur_title = ['4 neuron','30 neuron', '200 neuron', '1000 neuron', 'dcrrnn']
bins = 50
edge = 'red'
rang = [-2,1]
basin = 'Columbia'
title0 = [f"{basin}\n No SST, {x}" for x in neur_title]
title00 = [f"No SST, {x}" for x in neur_title]
title1 = [f"{basin}\n SST, {x}" for x in neur_title]
title11 = [f"SST, {x}" for x in neur_title]



x0 = pd.read_pickle(x[0])

#instantiate fig
fig = plt.figure(figsize=(15,2.5))
gs = gridspec.GridSpec(nrows=1,ncols=5)

x1 = (x0[x0['sst'] == 'False'])


for i in range(5):
    x2 = x1[x1['nn_hid_lay_size'] == value[i]]
    x3 = x2['nse_test']
    fig.add_subplot(gs[i])
    
    plt.hist(np.float32(x3),bins=bins,fill=False,edgecolor='green',range=rang);
    if i == 0:
        plt.xlabel('NSE')
        plt.title(title0[i])
    else:
        plt.title(title00[i])
plt.savefig('./plots/columbia_NOsst.png',bbox_inches='tight')

x4 = (x0[x0['sst'] == 'True'])

fig = plt.figure(figsize=(15,2.5))
gs = gridspec.GridSpec(nrows=1,ncols=5)
for i in range(5):
    x5 = x4[x4['nn_hid_lay_size'] == value[i]]
    x6 = x5['nse_test']
    fig.add_subplot(gs[i])
    plt.title(title1[i])
    plt.hist(np.float32(x6),bins=bins,fill=False,edgecolor='blue',range=rang);
    if i == 0:
        plt.xlabel('NSE')
        plt.title(title1[i])
    else:
        plt.title(title11[i])
plt.savefig('./plots/columbia_sst.png',bbox_inches='tight')
```
