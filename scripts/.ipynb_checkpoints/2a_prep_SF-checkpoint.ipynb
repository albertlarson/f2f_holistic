```python
def e(x):
    return enumerate(x)

from glob import glob as g
```

```python
colum = g('../data/SF/c*')
for x in sorted(colum):
    print(x)
```

```python
import pandas as pd
```

```python
import numpy as np
for idx,i in e(colum):
    y = []
    with open(i) as f:
        for x in f:
            y.append(x.strip())
    # for xdx,x in enumerate(y):
    #     print(xdx,x)

    z = y[37:253]
    a = []
    for x in z:
        a.append(x.split('\t')[4:])

    if idx == 0:
        b = a.copy()
        b = np.asarray(b)
        
    else:
        a = np.asarray(a)
        c = np.expand_dims(a[:,2],1)
        b = np.hstack((b,c))


c = pd.DataFrame(b,columns=['year',
                            'month',
                            colum[0][20:-4],
                            colum[1][20:-4],
                            colum[2][20:-4],
                            colum[3][20:-4],
                            colum[4][20:-4]
                           ]
                )
c
```
